# Glucose cell-lifelines
This notebook uses Monte Carlo simulation to track the location of a single cell in a bioreactor compartment model and links the location to a glucose concentration and glucose uptake rate.

## 1. Loading modules

```python
from cmodels.cc_cmodel import cc_cmodel
from cmodels.cc_cmodel_sol import cc_cmodel_sol
from mmodels.monod_bb.cc_monod_bb import cc_monod_bb
from cell_tracking.cell_tracking_fast import cell_tracking_fast

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Compartment model solution and cell tracking

### 2.1. STR (LS)
LS = Low setting. Refers to the stirring and aeration setting of the bioreactor

#### 2.1.1. Definition and solution

```python
# Loading the compartment model from excel definiton file
STR_LS_GNR = cc_cmodel(excel_path='cmodels/cmodel_files/RTD_LS/9_RTD_LS.xlsx')

# Defining metabolic model parameters for the Monod kinetic model
# Complex case means that the metabolic model is set up to function in combination with the more complex compartment model by Gisela Nadal Rey
# Metabolic rate equations of the model are:
# qS = qS_max*S/(S+Ks)*O/(O+Ko) | [gS/gX/h]   | Specific substrate upteake rate
# qO = qS*Yso                   | [mmol/gX/h] | specific oxygen uptake rate

complex_mmodel = cc_monod_bb(
    strain_id='DDB7',
    strain_description='WT E. coli'
)

complex_mmodel.define_strain_params(
    qS_max = 0.8,
    Ysx = 0.5,
    Ks = 0.05,
    Yso = 13,
    Ko = 0.002
)


# Initial values for substrate, oxygen and acetate. 
# Note: actete in this case is redundant, since the model does not consider acetate formation. Related rates and concentration will be 0.

y0_1d = [
    0.0, #S 
    0.2095, #O
    0 #A
    ]

# multiply initial values with the number of compartments
y0 = y0_1d*len(STR_LS_GNR.CompVolumes)

# simulation time to reach a steady state of the gradients
t_eval=np.linspace(0,1,10001)

# Biomass concentration X (stays constant)
X = 20 
# Growth rate for the exponential fed-batch
mu_set = 0.05
# Liquid volume in the whole reactor (compartment volumes minus the gas holdup (GH) in each compartment)
VL = (STR_LS_GNR.CompVolumes.CompVol - STR_LS_GNR.CompVolumes.CompVol*STR_LS_GNR.GH.GH).sum()
# Biomass yield (needed to calcualte the glucose feed rate Fs)
Ysx = complex_mmodel.Ysx
# Glucose feed rate with the unit [gS/h]
Fs_feed = mu_set/Ysx*X*VL*1000
# compartment model instance
cmodel = STR_LS_GNR

sol = cc_cmodel_sol(
    y0 = y0,
    t_eval = t_eval, 
    mmodel = complex_mmodel, 
    cmodel = cmodel, 
    X = X, 
    Fs_feed = Fs_feed
)
```

#### 2.1.2. Compartment model solution
DataFrame with endpoint values.

```python
sol.sol_df.comp_id = sol.sol_df.comp_id.astype('str')
sol_df_copy = sol.sol_df.copy()
sol_df_copy["qS/qSmax"] = sol_df_copy["qS [g/g/h]"]/0.8
sol_df_copy
```

#### 2.1.3. Cell tracking and coupling to compartment model solution

```python
def couple_cmodel_solution(cmodel_sol,cell_tracks,variable_str):

    cell_tracks_array = cell_tracks.cell_tracks
    compartment_model_solution = np.asarray(cmodel_sol.sol_df[variable_str])
    cell_lifelines = np.empty(shape=(cell_tracks_array.shape))

    for cell, lifeline in enumerate(cell_tracks_array):
        for step, comp in enumerate(lifeline):
            
            cell_lifelines[cell,step] = compartment_model_solution[int(comp)-1]
#         print(cell)
            
    return cell_lifelines
```

##### Saving timesteps

```python
# cell tracking
VL_array = np.array(STR_LS_GNR.CompVolumes.CompVol - STR_LS_GNR.CompVolumes.CompVol*STR_LS_GNR.GH.GH)
CompFlows = np.array(STR_LS_GNR.CompFlows)*3600
t_eval = np.linspace(0,2,240001)

ct = cell_tracking_fast(
    VL=VL_array,
    CompFlows = CompFlows,
    t_eval = t_eval,
    n_parcels = 1000
)

np.save(file='../data/cell_lifelines/time_steps_cell_lifelines', arr=ct.t_eval)
```

##### Perform the following: (1) calcualte cell tracking (2) couple to compartment model solution, (3) save relevant lifelines as npy files

```python
# cell tracking
VL_array = np.array(STR_LS_GNR.CompVolumes.CompVol - STR_LS_GNR.CompVolumes.CompVol*STR_LS_GNR.GH.GH)
CompFlows = np.array(STR_LS_GNR.CompFlows)*3600
t_eval = np.linspace(0,2,240001)

ct1 = cell_tracking_fast(
    VL=VL_array,
    CompFlows = CompFlows,
    t_eval = t_eval,
    n_parcels = 1000
)

ct1.calculate_cell_tracks()

#solution coupling
cell_lifelines = couple_cmodel_solution(sol,ct1,'qS [g/g/h]')
np.save(file='../data/cell_lifelines/qS_cell_lifelines1', arr=cell_lifelines)

cell_lifelines = couple_cmodel_solution(sol,ct1,'comp_id')
np.save(file='../data/cell_lifelines/comp_id_cell_lifelines1', arr=cell_lifelines)

# delete objects
del cell_lifelines
del ct1
```
